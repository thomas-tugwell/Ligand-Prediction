```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D

import statsmodels.api as sm
```

```python
df = pd.read_excel('PdPPe_Corr_master.xlsx', sheet_name='nonlinear_contour')
```

```python
x_columns = ['z0','E','S','E2','S2','ES','E3','S3','ES2','SE2']
y = df['Log_Fluor']
```

```python
def get_stats():
    x = df[x_columns]
    results = sm.OLS(y,x).fit()
    print(results.summary())

get_stats()
```
