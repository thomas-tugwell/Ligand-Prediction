```python
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
ligands[ligands.isnull().any(axis=1)]
```

```python
X = ligands.drop(['koide ID', 'log_fluor_expr'], axis=1)
```

```python
y = ligands['log_fluor_expr']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forward_feature_selection = SequentialFeatureSelector(RandomForestRegressor(n_jobs=-1),
                                                     k_features=6,
                                                     forward=True,
                                                     floating=False,
                                                     verbose=2,
                                                      scoring = "r2",
                                                     cv=5).fit(X_train, y_train)
```

```python
forward_feature_selection.k_feature_names_
```

```python
pd.DataFrame.from_dict(forward_feature_selection.get_metric_dict()).T
```
